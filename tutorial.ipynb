```python
estados = catalogos["Catálogo de ENTIDADES"]
```

```python
pd.merge(estados,confirmados,how="right",left_on="CLAVE_ENTIDAD",right_on="ENTIDAD_RES")
```

```python
confirmados_ampliado = pd.merge(estados,confirmados,how="right",left_on="CLAVE_ENTIDAD",right_on="ENTIDAD_RES")
```

```python
confirmados_ampliado["ENTIDAD_FEDERATIVA"].value_counts()
```

```python
confirmados_ampliado["ENTIDAD_FEDERATIVA"].value_counts().to_frame("Casos confirmados")
```

```python
confirmados_ampliado["ENTIDAD_FEDERATIVA"].value_counts().to_frame("Casos confirmados").reset_index()
```

```python
confirmados_ampliado["ENTIDAD_FEDERATIVA"].value_counts().to_frame("Casos confirmados").reset_index().rename(columns={"index":"Estado"})
```

```python
confirmados_estado = confirmados_ampliado["ENTIDAD_FEDERATIVA"].value_counts().to_frame("Casos confirmados").reset_index().rename(columns={"index":"Estado"})
```

```python
p = bpl.figure(x_range=confirmados_estado["Estado"], plot_width=800, plot_height=600,title="Casos confirmados por entidad federativa de residencia",y_range=(0,58000))
```

```python
p.vbar(x=confirmados_estado["Estado"],bottom=0,top=confirmados_estado["Casos confirmados"], width=0.8, color="#c00a56")
```

```python
p.text(x=confirmados_estado["Estado"],y=confirmados_estado["Casos confirmados"],text=confirmados_estado["Casos confirmados"],angle=3.1416/2,text_baseline="middle",y_offset=-5)
```

```python
p.xaxis.major_label_orientation = 3.1416/2
```

```python
p.yaxis.axis_label = 'Casos confirmados'
```

```python
p.grid.visible = False
```

```python
bpl.show(p)
```
